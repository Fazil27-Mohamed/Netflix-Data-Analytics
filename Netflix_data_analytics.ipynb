```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df['country'].fillna("Unknown", inplace=True)
df['director'].fillna("Unknown", inplace=True)
```

```python
type_counts = df['type'].value_counts()
type_counts
```

```python
type_counts.plot(kind='bar', title="Movies vs TV Shows")
plt.show()
```

```python
year_counts = df['release_year'].value_counts().sort_index()
year_counts
```

```python
year_counts.plot(kind='line', title="Content Growth Over Years")
plt.show()
```

```python
country_counts = df['country'].value_counts().head(10)
country_counts
```

```python
country_counts.plot(kind='barh', title="Top 10 Countries")
plt.show()
```

```python
genre_counts = df['listed_in'].str.split(',').explode().value_counts().head(10)
genre_counts
```

```python
genre_counts.plot(kind='bar', title="Top Genres")
plt.show()
```
